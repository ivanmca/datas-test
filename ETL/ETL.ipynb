```python
import pandas as pd
import sqlalchemy as db
from sqlalchemy import text
```

```python
engine = db.create_engine("mysql://root:root@127.0.0.1:3310/res_province_transact_db")
conn = engine.connect()
```

```python
query = """
SELECT st.id, pr.name AS pais, st.name AS direccion FROM res_province AS pr
INNER JOIN street AS st ON pr.id = st.res_province_id
"""
```

```python
street_data = pd.read_sql(query, conn)
# print(street_data)
street_data["id"] = street_data["id"].astype('int')
street_data
```
