```python
import pandas as pd
import sqlalchemy as db
from sqlalchemy import text
```

```python
engine = db.create_engine("mysql://root:root@127.0.0.1:3310/res_province_transact_db")
conn = engine.connect()
```

```python
query = """
SELECT st.id, pr.country_id, pr.name AS pais, st.name AS direccion FROM res_province AS pr
INNER JOIN street AS st ON pr.id = st.res_province_id
"""
```

```python
street_data = pd.read_sql(query, conn)
# print(street_data)
street_data["id"] = street_data["id"].astype('int')
street_data
```

```python
res_country=pd.read_csv("./data/res_country.csv")
res_country['currency_id'] = res_country['currency_id'].astype('int')
res_country.rename(columns={'name': 'pais', 'currency_id' : 'country_id'}, inplace=True)
res_country
```

```python
street_data = pd.merge(res_country, street_data, left_on='country_id', right_on='country_id')
street_data
```
